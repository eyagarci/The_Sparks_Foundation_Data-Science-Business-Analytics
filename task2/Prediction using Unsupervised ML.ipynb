# Prediction using Unsupervised ML

In this notebook, I tried to predict the optimum number of clusters and represent it visually from the given ‘Iris’ dataset using K-means clustering.



```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
```

```python
# load data
df=pd.read_csv('Iris.csv')
df.drop(['Id'],axis=1,inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
# label Encoding
le=LabelEncoder()
df['Species']=le.fit_transform(df['Species'])
df['Species'].value_counts()
```

```python
# compare PetalLengthCm vs PetalWidthCm
plt.scatter(df['PetalLengthCm'],df['PetalWidthCm'],c=df.Species.values)
```

```python
sns.pairplot(df)
```

```python
df=df.iloc[:,[0,1,2,3]].values
```

```python
# find the optimum number of clusters for k-means classification
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(df)
    # inertia method returns wcss for that model
    wcss.append(kmeans.inertia_)
    
wcss
```

```python
plt.figure(figsize=(10,5))
sns.set(style='whitegrid')
sns.lineplot(range(1, 11), wcss,marker='o',color='red')
plt.title('The Elbow Method', fontsize=20)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# init kmeans algorithm k=3
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 5)
y_kmeans = kmeans.fit_predict(df)
y_kmeans
```

```python
# display clusters
fig = plt.figure(figsize=(10, 7))
plt.title('Clusters with Centroids',fontweight ='bold', fontsize=20)
plt.scatter(df[y_kmeans == 0, 2], df[y_kmeans == 0, 3], s = 100, c = 'seagreen', label = 'Iris-versicolour')
plt.scatter(df[y_kmeans == 1, 2], df[y_kmeans == 1, 3], s = 100, c = 'purple', label = 'Iris-setosa')
plt.scatter(df[y_kmeans == 2, 2], df[y_kmeans == 2, 3],s = 100, c = 'yellow', label = 'Iris-virginica')
plt.scatter(kmeans.cluster_centers_[:, 2], kmeans.cluster_centers_[:,3], s = 300, c = 'red',marker='o', 
            label = 'Centroids')
plt.title('Iris Flower Clusters', fontsize=25)
plt.ylabel('Petal Width in cm')
plt.xlabel('Petal Length in cm')
plt.legend()
```
